### Notice the csv file is a clean file type the reason we still have our total number of rows the same

```python
advert_df = advert_df.dropna()
advert_df
```

```python
advert_df.describe()
```

### Testing Phase

```python
from sklearn.model_selection import train_test_split

X = advert_df[['TV']]
Y = advert_df['Sales']

X_train, X_test, Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
X_train.sample(5)
```

```python
Y_train.sample(5)
```

### Training the Algorithm, we have to import LinearRegression 

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression(normalize=True).fit(X_train,Y_train)
```

```python
('Training score: ', linear_model.score(X_train, Y_train)) 
```

### We can then predict variables for the testing, notice we have to import r2 from sklearn.metrics

```python
Y_pred = linear_model.predict(X_test) #y_test
```

```python
from sklearn.metrics import r2_score
print('Testing score: ', r2_score(Y_test,Y_pred))
```

```python
fig, ax = plt.subplots(figsize = (12,8))

plt.scatter(X_test,Y_test)
plt.plot(X_test, Y_pred, color ='r')

plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```
